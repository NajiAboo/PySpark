```python
from pyspark.sql.types import StructType, StructField, IntegerType, DoubleType, StringType
```

```python
dict_scheam= [
    StructField("Player Name", StringType(), True),
    StructField("Rounds", DoubleType(), True),
    StructField("Fairway Percentage", DoubleType(), True)
]
```

```python
schema = StructType(fields=dict_scheam)
```

```python
df = spark.read.csv("dataset/pgaTourData.csv", schema=schema)
```
